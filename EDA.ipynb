```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('synthetic_rumors_dataset.csv')

#bar plot 
plt.figure(figsize=(8, 6))
df['Label'].value_counts().plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Bar plot Distribution of Rumors and Facts')
plt.xlabel('Label')
plt.ylabel('Count')

plt.show()
```

```python
# Scatter plot 1 
x = df['Text'] 
y = df['Label']
plt.scatter(x, y)
plt.xlabel('Text')
plt.ylabel('Label')
plt.title('Scatter Plot With Text')
plt.show()
```

```python
# Scatter plot 2
plt.scatter(df.index, df['Label'])
plt.xlabel('Index')
plt.ylabel('Label')
plt.title('Scatter Plot with Index Values')
plt.show()
```

```python
# Displaying  the head dataframe
print(df.head())
```

```python
# Displaying the tail  dataframe
print(df.tail())
```

```python
# Plot index vs label
plt.plot(df.index, df['Label'])
plt.xlabel('Index')
plt.ylabel('Label')
plt.title('Plot of Index vs Label')
plt.show()
```

```python
# Pie chart distribution
plt.pie(df['Label'].value_counts(), labels=df['Label'].unique(), autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Pie Chart Distribution')
plt.show()
```
